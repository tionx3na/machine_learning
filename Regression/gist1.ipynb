Diabetes dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model, model_selection
```

```python
#Loading diabetes dataset from scikit-learn's dataset
diabetes = datasets.load_diabetes()
print(diabetes.DESCR)
```

```python
table = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
table.head(10)
```

As per dataset documentation: "Each of these 10 feature variables have been mean centered and scaled by the standard deviation times n_samples"

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0]) # printing out the first array with 10 elements
```

The parameter return_X_y in load_diabetes returns (data,target) instead of a Bunch object. Bunch object is a dictionary like object with attributes like,
data, target, feature_names, frame etc.

We can observe the shape of the dataset, (442,10) that is, there are 442 arrays consisting of 10 elements each, where each each of these elements are the features of the dataset.

```python
bmi = X[:,np.newaxis,2]
print(bmi.shape)
```

```python
X_train ,X_test, y_train, y_test = model_selection.train_test_split(bmi,y,test_size=0.33)
print(X_train.shape,X_test.shape,y_train.shape, y_test.shape)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color ='red')
plt.plot(X_test, y_pred, color='black', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('Diabetes progression against BMI')
plt.show()
```

Here we see how a straight line is obtained by linear regression by calculating the minimum distance between the datapoints of bmi. The above picture shows the progression of diabetes with increase or decrease in BMI.

```python
bp = X[:,np.newaxis,2]
X_train ,X_test, y_train, y_test = model_selection.train_test_split(bp,y,test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
plt.scatter(X_test, y_test, color ='green')
plt.plot(X_test, y_pred, color='black', linewidth=3)
plt.xlabel('Scaled Blood pressure')
plt.ylabel('Disease Progression')
plt.title('Diabetes progression against BP')
plt.show()
```
